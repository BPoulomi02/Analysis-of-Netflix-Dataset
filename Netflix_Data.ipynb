```python
import pandas as pd
dataset=pd.read_csv("/content/archive (4) (1).zip")
```

```python
import matplotlib.pyplot as plt
# Extract relevant columns
dataset[['Title', 'Runtime']]

# Convert duration to numeric value (assuming it's in minutes)
dataset['Runtime'] = pd.to_numeric(dataset['Runtime'])

# Plot the data to observe trends
plt.figure(figsize=(10, 6))
plt.plot(dataset['Runtime'], marker='o', linestyle='-')
plt.title('Trend in Content Duration on Netflix')
plt.xlabel('Index')
plt.ylabel('Duration (minutes)')
plt.grid(True)
plt.show()
```

```python
# Split genres in the 'listed_in' column
Genres=dataset['Genre'].str.split(', ')

# Flatten the list of lists
Genres = [Genre for sublist in Genres.dropna() for Genre in sublist]

# Create a DataFrame to count the occurrence of each genre
Genres_counts = pd.Series(Genres).value_counts()

# Plot the top 10 prevalent genres
top_Genres = Genres_counts.head(10)
plt.figure(figsize=(10, 6))
top_Genres.plot(kind='bar', color='skyblue')
plt.title('Top 10 Most Prevalent Genres on Netflix')
plt.xlabel('Genre')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Extract relevant columns
country_counts = dataset['Languages'].value_counts()

# Plot a pie chart for the distribution of content by country
plt.figure(figsize=(10, 8))
country_counts.head(10).plot(kind='pie', autopct='%1.1f%%', colors=plt.cm.tab20.colors)
plt.title('Distribution of Content by Languages on Netflix')
plt.ylabel('')  # Remove the ylabel
plt.show()


```

```python
# Count the occurrences of each type of content
content_counts = dataset['Series or Movie'].value_counts()

# Plot the distribution of content types
plt.figure(figsize=(6, 4))
content_counts.plot(kind='bar', color=['skyblue', 'orange'])
plt.title('Distribution of Content Types on Netflix')
plt.xlabel('Content Type')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()

```
